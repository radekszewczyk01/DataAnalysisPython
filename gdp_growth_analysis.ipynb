## Economic growth can be observed by calculating the relative change in GDP

```python
dataset['Calculated_Growth'] = (dataset['2023 [YR2023]'].astype(float) - dataset['2014 [YR2014]'].astype(float))/dataset['2014 [YR2014]'].astype(float)
dataset_sorted_calculated_growth = dataset.sort_values(by='Calculated_Growth', ascending=False)
dataset_sorted_calculated_growth.head(20)
```

```python
def plot_countrys_gdp_in_years(country, starting_year=1990, ending_year=2024):
    expected_years = [str(year) + ' [YR' + str(year) + ']' for year in range(starting_year, ending_year)]
    available_years = [year for year in expected_years if year in dataset.columns]

    countrys_data = dataset[dataset['Country Name'] == country][available_years]
    
    values = countrys_data.values.flatten()
    values = pd.to_numeric(values, errors='coerce')  # zamienia stringi na liczby, bdne wartoci na NaN
    # Wykres
    plt.figure(figsize=(12, 6))
    plt.plot(available_years, values, marker='o', linestyle='-', color='blue')
    
    plt.title('Warto sprzeday w {} (wybrane lata)'.format(country))
    plt.xlabel('Rok')
    plt.ylabel('Warto')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def calculate_growth_in_years(starting_year=1990, ending_year=2024):
    dataset['Calculated_Growth'] = (dataset[str(ending_year) + ' [YR' + str(ending_year) + ']'].astype(float) - dataset[str(starting_year) + ' [YR' + str(starting_year) + ']'].astype(float))/dataset[str(starting_year) + ' [YR' + str(starting_year) + ']'].astype(float)
    dataset_sorted_calculated_growth = dataset.sort_values(by='Calculated_Growth', ascending=False)
    return dataset_sorted_calculated_growth
```
