```python
dataset['Calculated_Growth'] = dataset['2023 [YR2023]'].astype(float) - dataset['1990 [YR1990]'].astype(float)
dataset_sorted_calculated_growth = dataset.sort_values(by='Calculated_Growth', ascending=False)
dataset_sorted_calculated_growth.head()
```

```python
def plot_countrys_gdp_in_years(country):
    expected_years = [str(year) + ' [YR' + str(year) + ']' for year in range(1990, 2024)]
    available_years = [year for year in expected_years if year in dataset.columns]

    countrys_data = dataset[dataset['Country Name'] == country][available_years]
    
    values = countrys_data.values.flatten()
    values = pd.to_numeric(values, errors='coerce')  # zamienia stringi na liczby, bdne wartoci na NaN
    # Wykres
    plt.figure(figsize=(12, 6))
    plt.plot(available_years, values, marker='o', linestyle='-', color='blue')
    
    plt.title('Warto sprzeday w {} (wybrane lata)'.format(country))
    plt.xlabel('Rok')
    plt.ylabel('Warto')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
